## REGEX

[Документация](https://docs.python.org/3/library/re.html)

```python
import re
```

### re.match(pattern, string):
Этот метод ищет по заданному шаблону в начале строки. Например, если мы вызовем метод match() на строке «Мама мыла раму» с шаблоном «Мама», то он завершится успешно. Однако если мы будем искать «мыла», то результат будет отрицательный. Давайте посмотрим на его работу:

```python
result = re.match(r'Мама', 'Мама мыла раму')
print(result)
```

```python
result = re.match(r'Мама', 'Мама мыла раму Мама')
print(result.group(0))
```

```python
result = re.match(r'мыла', 'Мама мыла раму Мама')
print(result)
```

### re.search(pattern, string):
Этот метод похож на match(), но он ищет не только в начале строки. В отличие от предыдущего, search() вернет объект, если мы попытаемся найти «мыла».
Метод search() ищет по всей строке, но возвращает только первое найденное совпадение.

```python
result = re.search(r'мыла', 'Мама мыла раму Мама')
print(result)
```

### re.findall(pattern, string):
Этот метод возвращает список всех найденных совпадений. У метода findall() нет ограничений на поиск в начале или конце строки. Если мы будем искать «Мама» в нашей строке, он вернет все вхождения «Мама». Для поиска рекомендуется использовать именно findall(), так как он может работать и как re.search(), и как re.match().

```python
result = re.findall(r'Мама', 'Мама мыла раму Мама')
print(result)
```

### re.split(pattern, string, [maxsplit=0]):
Этот метод разделяет строку по заданному шаблону.

```python
result = re.split(r' ', 'Мама мыла раму Мама')
print (result)
```

### re.sub(pattern, repl, string):
Этот метод ищет шаблон в строке и заменяет его на указанную подстроку. Если шаблон не найден, строка остается неизменной.

```python
result = re.sub(r'Мама', 'Кошка', 'Мама мыла раму Мама')
print(result)
```

### re.compile(pattern, repl, string):
Мы можем собрать регулярное выражение в отдельный объект, который может быть использован для поиска. Это также избавляет от переписывания одного и того же выражения.

```python
pattern = re.compile('Мама')
result = pattern.findall('Мама мыла раму Мама')
print(result)
```

```python
result2 = pattern.sub('Кошка','Мама мыла раму Мама')
print(result2)
```
